```python
#Importando a biblioteca
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Lendo os dados da planilha execel e armazenamos na varíavel "dados", que conterá o resultado da leitura    
dados = pd.read_excel("Vendas.xlsx")
#Exibindo os dados do dataflame, os dados que desjamos efetuar a análise
display(dados)
```

```python
dados.columns
```

```python
#Consultando linhas com valores faltantes
dados.isnull().sum()
```

```python
#Extraindo o ano,  mês e o dia e acrecentando as colunas existentes 
dados["ano_venda"], dados["mes_venda"], dados["dia_venda"] = (dados["Data"].dt.year,dados["Data"].dt.month, dados["Data"].dt.day)
dados
```

```python
#aturamento total 
Faturamento_total = dados["Valor Final"].sum()
Faturamento_total
```

```python
#Faturamento por loja
Faturamento_total_por_loja = dados[['ID Loja', 'Valor Final']].groupby('ID Loja').sum().sort_values('Valor Final', ascending=False)
display(Faturamento_total_por_loja)
```

```python
#Adicionando um título e alterando o nome dos eixos
plt.figure(figsize=(20,5))
dados.groupby(dados["ID Loja"])["Valor Final"].sum().plot(
    title="Total Faturamento x Loja")
plt.xlabel("Loja")
plt.ylabel("Faturamento")
plt.legend()
```
